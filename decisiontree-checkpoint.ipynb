```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
delivery_data= pd.read_csv('C:/data/delivery_data.csv', encoding='utf-8')
delivery_data.head()
```

```python
delivery_data['city'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False)

ohe.fit(delivery_data[['city']])

city_list = list(delivery_data['city'].unique())

city_encoding = pd.DataFrame(ohe.transform(delivery_data[['city']]), columns=city_list, dtype='int64')
city_encoding
```

```python
test = pd.concat([delivery_data,city_encoding],axis=1)
```

```python
city_list.append('city')
city_list.append('humidity')
city_list.append('precipitation')
city_list.append('temp')
city_list.append('windspeed')
```

```python
#feature_names = ["city","humidity","precipitation","temp","windspeed"]
dfX = test[city_list]
dfy = test["sectors"]
dfy = dfy.astype('int')
dfX.tail()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
X_train, X_test, y_train, y_test = train_test_split(dfX, dfy, test_size=0.25, random_state=1)
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=5, min_samples_leaf=5).fit(X_train, y_train)
```
